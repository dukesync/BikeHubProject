### write to bronze layer

```python
for item in ingestion_config:
    print(f"Ingesting {item['source']} -> bronze.{item['table']}")

    # reading file
    df = (
    spark.read
        .option("header", True)
        .option("inferSchema", True)
        .csv(item["path"])
    )
    # write to bronze table
    ( df.write
        .mode("overwrite")
        .saveAsTable(f"bronze.{item['table']}")
    )  
```
