```python
# What is the total sales amount and number of orders by customer gender?
query = """
SELECT
  c.gender,
  SUM(f.sales_amount) AS total_sales_amount,
  COUNT(DISTINCT f.order_number) AS total_orders
FROM
  bike_data.gold.fact_sales f
JOIN
  bike_data.gold.dim_customer c
ON
  f.customer_sk = c.customer_id
GROUP BY
  c.gender
ORDER BY
  total_sales_amount DESC
"""
df = spark.sql(query)
display(df)
```
