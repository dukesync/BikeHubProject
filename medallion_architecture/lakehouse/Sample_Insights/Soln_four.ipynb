```python
# Product Line Revenue: What is total revenue and units sold by product_family and Subcategory, for currently active products only
query = """
SELECT 
    p.product_family,
    p.subcategory,
    SUM(f.quantity) AS total_units_sold,
    SUM(f.sales_amount) AS total_revenue,
    -- Calculate Average Selling Price (ASP) for extra insight
    ROUND(SUM(f.sales_amount) / SUM(f.quantity), 2) AS avg_selling_price
FROM bike_data.gold.fact_sales f
JOIN bike_data.gold.dim_products p ON f.product_sk = p.product_sk
WHERE p.end_date IS NULL -- Only active products
GROUP BY 1, 2
ORDER BY p.product_family ASC, total_revenue DESC;
"""
df = spark.sql(query)
```
